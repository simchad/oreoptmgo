```python
import pandas as pd
import numpy as np

df = pd.read_csv(filepath_or_buffer="C:/Users/simhc/Downloads/UserSubs.CSV")
entry = df['Entry'].tolist()
entry[0]
```

```python
import pyiptmnet.api as api
import pandas as pd


# Assuming you have a list of UniProt IDs
uniprot_ids = entry  # Add your 200+ UniProt IDs here

# Initialize an empty DataFrame to store the concatenated results
all_results = pd.DataFrame()

# Loop through each UniProt ID
for uniprot_id in uniprot_ids:
    try:
        # Call the get_substrates() function
        result = api.get_substrates(uniprot_id)
        # Convert the result to a DataFrame (assuming result is in a suitable format)
        result_df = pd.DataFrame(result)
        # Concatenate the result to the all_results DataFrame
        all_results = pd.concat([all_results, result_df], ignore_index=True)
    except Exception as e:
        # Print the error message and pass the UniProt ID
        print(f"Error with UniProt ID {uniprot_id}: {e}")
        continue

all_results.info
all_results.to_csv(path_or_buf="C:/Users/simhc/Downloads/all_results.CSV", encoding="utf-8", index=False)
```

```python
import csv

def extract_fasta_headers_to_csv(fasta_file, csv_file):
    headers = []
    with open(fasta_file, 'r') as file:
        for line in file:
            if line.startswith('>'):
                headers.append(line.strip())

    with open(csv_file, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Header'])
        for header in headers:
            writer.writerow([header])

# Example usage
fasta_file = "C:/Users/simhc/Downloads/idmapping_2024_08_21.fasta"
csv_file = "C:/Users/simhc/Downloads/RetrieveFastaHeader.CSV"
extract_fasta_headers_to_csv(fasta_file, csv_file)
print(f"Headers have been written to {csv_file}")
```

```python
import csv

def extract_fasta_headers_to_csv(fasta_file, csv_file):
    headers = []
    with open(fasta_file, 'r') as file:
        for line in file:
            if line.startswith('>'):
                header = line.strip()
                protein_id = header.split('|')[1]  # Extract protein identifier between '|'
                headers.append((protein_id, header))

    with open(csv_file, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Protein Identifier', 'Header'])
        for protein_id, header in headers:
            writer.writerow([protein_id, header])

# Example usage
fasta_file = "C:/Users/simhc/Downloads/idmapping_2024_08_21.fasta"
csv_file = "C:/Users/simhc/Downloads/RetrieveFastaHeader.CSV"
extract_fasta_headers_to_csv(fasta_file, csv_file)
print(f"Headers have been written to {csv_file}")
```
